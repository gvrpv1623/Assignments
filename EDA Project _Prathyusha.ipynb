```python
plt.figure(figsize=(12,5))
#the kdeplot of the rating column 
ax=sns.kdeplot(data=df['Rating'],shade=True,color='g')
#the y-axis column values to plot mean ,median ,mode ,25thpercentile,75th percentile
y=[0.0,0.025,0.050,0.075,0.1,0.125,0.150,0.175]
               
#to plot the mean line on the kdeplot
x=[df['Rating'].mean()]*8
plt.plot(x,y)

#to plot the median line on the kdeplot
x2=[df['Rating'].median()]*8
plt.plot(x2,y)

#to plot the mode line on the kdeplot
x3=[df['Rating'].mode()]*8
plt.plot(x3,y)
               
#to plot the 25th percentile line on the kdeplot               
x4=[np.percentile(df['Rating'],25)]*8
plt.plot(x4,y)
               
#to plot the 75th percentile line on the kdeplot
x5=[np.percentile(df['Rating'],75)]*8
plt.plot(x5,y)
               
#giving labels to the plotted lines
plt.legend(['mean','median','mode','25th percentile','75th percentile'],loc='upper right')
```

from the plot we can say that the data in rating is normally distributed

as we can see , the mean median lines are very close to each other

since the difference between the mean median of the data is very less ,skewness is almost not present

# 3. Inference from the aggregate sales across the branches

```python
#applying the groupby function on the basis of product line column and
#get the sum of the total column  for the columns where the gender condition is given (female ,male)
f=df[df['Gender']=='Female'].groupby('Product line')['Total'].sum().sort_values(ascending=False)
m=df[df['Gender']=='Male'].groupby('Product line')['Total'].sum().sort_values(ascending=False)

#Spending function to plot the spending on the basis of gender
def spending(f):
    
    #bargraph for the total made on each products
    bar=sns.barplot(x=f.index,y=f.values,palette='rainbow') 
    
    #to print the label on y-axis
    plt.ylabel("Amount Spent",fontsize=12)
    
    #to print the text on each bar
    for i in range(len(f)):
        plt.text(i,f.values[i],f.values[i].round(2),ha="center",va="bottom",fontsize='15')

        
plt.figure(figsize=(25,8))
#subplot to plot the two graphs in 1*2 matrix form
plt.subplot(1,2,1)
spending(f)#calling the sepending function for female
plt.subplot(1,2,2)
spending(m)#calling the spending function for male
plt.show()
```

the spending of males is high in Health and beauty

```python
#now in the city Naypyitaw we consider which products to focus on

#grouping on the basis of product and getting the total no.of items sold in each product in the supermarket 
#for the city Naypyitaw
x=df[df['City']=='Naypyitaw'].groupby('Product line')['Quantity'].sum()

print("no.of items sold for each product line in the Naypyitaw supermarket\n",x)
#to plot the bargraph
plt.figure(figsize=(20,10))
color=["silver",'turquoise',"lightcoral",'greenyellow']
plt.bar(x.index,x.values.round(2),color=color,ec="black")

#labels and titles to bargraph
plt.xlabel("Products",fontsize=18)
plt.ylabel("Total No.of items sold",fontsize=18)
plt.title("Total No.of items sold in each product",fontsize=20)

#to give grid look to the plot
plt.grid(True)

#to plot text on each bar in the plot
for i in range(len(x)):
    plt.text(i,x.values[i],x.values[i],ha="center",va="bottom",fontsize=12)
    
plt.show()
```

from the above graph it can be infered that 

i.products with less no.of sales that has to be focused on are(to improve their sales)

Home and Lifestyle products(less sales when compared to other product line sales)

Health and beauty products 

and next to that Sport and Travel

ii.products with more no.of sales that has to be focused on are(to manage with heavy sales)

Food and beverages(more sales when compared to other product sales)
