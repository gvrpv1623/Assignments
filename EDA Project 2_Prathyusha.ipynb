# Adding required Columns to the DataFrame

to perform the required anlysis some data columns have to be added to the Dataframe,they are

from the date column the following has to be extracted and add into the Dataframe

'hour' column to know the point of hour of the Sales

'week_day' column to know the name of the week in which the sales were made

```python
df['hour']=df['Date'].dt.hour
```

```python
df['week_day']=df['Date'].dt.day_name()
```

```python
df['month']=df['Date'].dt.month_name()
```

Adding 'part_of_day' column to know which part of the day (i.e,Morning,Afternoon,Evening,night) the sales were made

get_part_of_day function  returns the part of the day (i.e,Morning,Afternoon,Evening,night)

by taking input from the hour column

input:returns

5-12:morning

12-17:afternoon

17-22:evening

rest of the time considered as night

```python
def get_part_of_day(h):
    return (
        "morning"
        if 5<=h<12
        else"afternoon"
        if 12<=h<=17
        else "evening"
        if 17<h<=22
        else "night"
    )

#applying the get_part_of_day function to the dataframe over the column hour
#saving the values by adding part_of_day column into the dataframe

df['part_of_day']=df['hour'].apply(get_part_of_day)
```

Adding 'city' column to know the city in which the sales were made


```python
#using the split function over the comma',' to get the city name which is in the middle of the Purchase Address
#since there are two commas,the one after the first comma is the city name
#so used [1]
df['city'] = df['Purchase Address'].str.split(',', expand=True)[1]
```

Adding 'Sales' column by multiplying the 'Price Each' and 'Quantity Ordered' columns to know the amount of sales being made from a particular order

```python
#adding the sales column
df['Sales']=df['Price Each']*df['Quantity Ordered']
```

# 6.Finding the pair of products that are most sold together

getting the Product column values over the duplicated values in Order ID

since if a person brought items together then the order ID value remains the same

and that list is grouped over the order ID column and joining the values of Product that has the same Order ID with the ','

finally getting the first n rows from the repeated Product column obtained after joining

to put in a simple way grouping over Order ID for the duplicated row values

Products that are brought together are joined together

finding the unique values count of the joined products

and extracting the top n required Joined Products that are Products most sold together 

```python
#to extract the duplicate Order ID value rows into duplicates
duplicates=df[df.duplicated('Order ID',keep=False)].filter(['Order ID','Product'])

#applying the groupby function over Order ID column from the duplicated rows
#and joining the Product column values for the same  Order ID column values
df2=duplicates.groupby('Order ID')['Product'].apply(','.join).reset_index()
n=30

#to get the first 30 Products that are sold together
df2['Product'].value_counts()[:n]
```
